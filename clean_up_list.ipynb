### phone number formatting

```python


def format_phone(phone):
    phone = str(phone).replace("-", "")  
    
   
    if len(phone) == 11 and phone[0] == '1':
        phone = phone[1:]  # Remove the leading '1'
    
   
    if len(phone) == 10:
        
        return f"{phone[:3]}-{phone[3:6]}-{phone[6:10]}"
    else:
       
        return phone


dataframe['phone'] = dataframe['phone'].apply(format_phone)


#dataframe['phone'].to_string()

dataframe.loc[286:295]

```
